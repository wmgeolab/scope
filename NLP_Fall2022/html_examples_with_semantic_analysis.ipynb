**Keyword Extraction** [https://towardsdatascience.com/keyword-extraction-with-bert-724efca412ea]

```python
china_1 ## the text i'm  going to be using here as a first test.
```

```python
from sklearn.feature_extraction.text import CountVectorizer

n_gram_range = (1, 1) # will need to consider changing this.
stop_words = "english"

# Extract candidate words/phrases
count = CountVectorizer(ngram_range=n_gram_range, stop_words=stop_words).fit([china_1])
candidates = count.get_feature_names_out()
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('distilbert-base-nli-mean-tokens')
doc_embedding = model.encode([china_1])
candidate_embeddings = model.encode(candidates)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

top_n = 5
distances = cosine_similarity(doc_embedding, candidate_embeddings)
keywords = [candidates[index] for index in distances.argsort()[0][-top_n:]]
```

```python
## Semantic Search
```

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('multi-qa-MiniLM-L6-cos-v1')
```

```python
test_sentences = china_1.split(".\n")
```

```python
test_sentences.remove("")
```

```python
import torch
```

```python
query_embedding = model.encode("people involved")
passage_embedding = model.encode(test_sentences)
```

```python
top_k = 3
```

```python
cos_scores = util.cos_sim(query_embedding, passage_embedding)[0]
top_results = torch.topk(cos_scores, k=top_k)
```

```python
for score, idx in zip(top_results[0], top_results[1]):
    print(test_sentences[idx], "(Score: {:.4f})".format(score))
```

```python
query_embedding = model.encode("geographic area")
passage_embedding = model.encode(test_sentences)
top_k = 3
cos_scores = util.cos_sim(query_embedding, passage_embedding)[0]
top_results = torch.topk(cos_scores, k=top_k)
for score, idx in zip(top_results[0], top_results[1]):
    print(test_sentences[idx], "(Score: {:.4f})".format(score))
```
